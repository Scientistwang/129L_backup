```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# DC + 40 Hz + 150Hz
t = np.linspace(0.,0.5,500)
x = 2 + 3*np.sin(40*(2*math.pi)*t) - 8*np.cos(150*(2*math.pi)*t+0.5)
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(t,x)
```

```python
fft = np.fft.fft(x)
T = t[1]-t[0]
N = len(x)
f = np.linspace(0, 1 / T, N)
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(f, abs(fft)/N)
```

```python
fig3, ax3 = plt.subplots()
ax3.plot(f, np.angle(fft, deg=True))
```
